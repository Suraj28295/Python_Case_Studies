```python
#Q.1)How many attacks happened in India?

india_attack_count = df[df.country_txt=='India'].shape[0]

print("Number of attacks in india:"+str(india_attack_count))
```

```python
#Q.2)How many attacks happened in India and upto 3 people were killed?

india_more_than_3 = df[(df.country_txt=='India') & (df.nkill > 3)]

print("Number of attacks where killed is more than 3:"+str(india_more_than_3.shape[0]))
```

```python
#Q.4)In a single terror incident in India, find out top 5 cities by number killed

top5_nkill_india = df[df.country_txt=='India'].sort_index(by='nkill',ascending=False).head(5)


print(top5_nkill_india)
```

```python
#Q.5)	In a single terror incident in India, find out top 5 cities by number killed and wounded

top5_killed_wounded = df[df.country_txt=='India'].sort_index(by=['nkill','nwound'],ascending=False).head(5)

print(top5_killed_wounded[['nkill','nwound']])
```

```python
#Q.6) How many attacks were successful that were suicide attacks?

print("Successful attacks "+str(df[['suicide']].sum()))
```

```python
#Q.7) label all the incidents where the number killed was more than 5 as severe. 

#df['incidents']= np.where(df['nkill']>5,'severe','not severe')
df['incidents']= ['severe' if i > 5 else 'not severe' for i in df['nkill']]
```

```python
#Q.8) write a function to label an incident that was both successful and suicidal

#df['Success_and_suicidal']= np.where((df['success']==1) & (df['suicide']==1),1,0)

nrows=list(range(0,df.shape[0]))
df['Success_and_suicidal']= [1 if df.iloc[i].suicide==1 and df.iloc[i].success==1  else 0 for i in nrows]
 
```

```python
#Q.9)	Create a new category representing if the incident occured in Afghanistan, Pakistan or India as one level of the category and all the other countries as another level

df['levels'] =['Asia' if df.iloc[i].country_txt in ['Afghanistan','Pakistan','India']  else 'ROW' for i in nrows]
```

```python
#Q.10) How many incidents happened in Af-Pak-India vs ROW?
print("No. of Incidents in Afghanisthan, Pakistan and India is: ", len(df.query("levels=='Asia'")))
len(df.query("levels=='ROW'"))
print(round(len(df.query("levels=='Asia'"))*100/len(df.query("levels=='ROW'")),3))
```

```python
#Q.11)	List the number of suicides attacks and average kills by Af-Pak-India vs ROW. Rename columns in the output as Average_Kills and Number_Incidents.

print(df.query("suicide == 1").groupby('levels')[['nkill']].agg(['count','mean']).rename(columns = {'count' : "Number_Incidents",'mean':"Average_kills"}))
```
